```python
import os 
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
import numpy as np
if not hasattr(np, 'bool8'):
    np.bool8 = np.bool_
```

```python
env = gym.make("Swimmer-v5", render_mode='human')
```

```python
env.action_space
```

```python
env.observation_space
```

```python
episodes = 5
for episode in range(episodes):
    state = env.reset()
    done = False
    truncated = False
    score = 0

    while not done and not truncated:
        action = env.action_space.sample()
        n_state, reward, done, truncated, info = env.step(action)
        score += reward
    print(f"Episode: {episode+1} Score:{score}")
env.close()
```

```python
train_env = gym.make("Swimmer-v5")
train_env = DummyVecEnv([lambda : train_env])
model = PPO('MlpPolicy', train_env, verbose=1)
```

```python
model.learn(total_timesteps=5000000)
```

```python
PPO_path = os.path.join('Saved Models', 'PPO_5M')
model.save(PPO_path)
```

```python
del model
```

```python
test_env = gym.make("Swimmer-v5", render_mode='human')
```

```python
model = PPO.load(PPO_path, env=test_env)
```

```python
evaluate_policy(model, test_env, n_eval_episodes=10, render=False)
```

```python
env.close()
```

```python
episodes = 5
for episode in range(episodes):
    obs, _ = test_env.reset()
    done = False
    truncated = False
    score = 0

    while not done and not truncated:
        action = model.predict(obs)
        act = np.array(action[0])
        action = tuple(act)
        obs, reward, done, truncated, info = test_env.step(action)
        score += reward
    print(f"Episode: {episode+1} Score:{score}")
test_env.close()
```
